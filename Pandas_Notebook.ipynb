# Creating

```python
import pandas as pd
```

**DataFrame**

```python
# Column by Column Method
df1 = pd.DataFrame(
    {"StudentID" : ["C1100039"]}
)
print(df1)

df1 = pd.DataFrame(
    {"StudentID" : "C1100039"},
    index = [0]
)
print(df1)

# Row by Row Method
df1 = pd.DataFrame([
    {"StudentID" : "C1100039"}
])
print(df1)
```

```python
# Column by Column Method
df2 = pd.DataFrame({
    'Student' : ['C2211', 'C2222', 'C3333'],
    'Name' : ['A', 'B', 'C'],
    'Marks' : [100, 91, 87]
},
index=[1,2,3])
print(df2)

# Row by Row Method
df2 = pd.DataFrame([
    {'StudentID': 'C2211', 'Name': 'A', 'Marks': 100},
    {'StudentID': 'C1111', 'Name': 'B', 'Marks': 92},
    {'StudentID': 'C3333', 'Name': 'C', 'Marks': 85}
])
print('\n', df2)
df2
```

```python
# Display the first 5 rows to quickly inspect the data format
df2.head()
```

```python
# Get the dimensions of the DataFrame (number of rows and columns)
df2.shape
```

```python
# View a summary of data types and check for any missing/null values
df2.info()
```

```python
df2.to_csv('xyz.csv', index=False)
df2.to_excel('xyz.xlsx')
df2.to_json('xyz.json')
```

**Series**

```python
df3 = pd.Series([1,2,3,4,5])
print(df3)

df3 = pd.Series([50, 62, 77], index= ['2024 Sales', '2025 Sales', '2026 Sales'], name='Product A')
print('\n',df3)
```
