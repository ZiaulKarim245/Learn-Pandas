```python
# Column by Column Method
df1 = pd.DataFrame(
    {"StudentID" : ["C1100039"]}
)
print(df1)

df1 = pd.DataFrame(
    {"StudentID" : "C1100039"},
    index = [0]
)
print(df1)

# Row by Row Method
df1 = pd.DataFrame([
    {"StudentID" : "C1100039"}
])
print(df1)
```
